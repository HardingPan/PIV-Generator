### The config for different flows 

```python
import numpy as np
import matplotlib.pyplot as plt

from config import config
from generator import Generator
from plot import plot_field

cfg = config()
pig = Generator(cfg)
```

```python
def flowtest(cfg):
    pig = Generator(cfg)
    img1, img2, ut, vt = pig.compute()
    
    # show the velocity field
    x, y = np.meshgrid(np.arange(ut.shape[0]), np.arange(ut.shape[1]), indexing="ij")
    amp = np.sqrt(ut**2+vt**2)
    fig = plot_field(x,y,ut,vt,bkg=amp,cmap=None,figsize=(8,3.2))


# **** "uniform_flow", "sine_flow", "lamboseen_flow", "cellular_flow" ****
# uniform flow
cfg.flow = "uniform_flow" 
cfg.uniform_flow = {'c_x':4.75, 'c_y':0.} # uniform_flow
flowtest(cfg)
plt.title(cfg.flow)

# sine_flow
cfg.flow = "sine_flow" # "uniform_flow", "sine_flow", "lamboseen_flow", "cellular_flow"
cfg.sine_flow = {'a':5, 'px':128, 'vmax':4.75} 
flowtest(cfg)
plt.title(cfg.flow)

# lamboseen_flow
cfg.flow = "lamboseen_flow" 
cfg.lamboseen_flow['gamma'] = 5e3
flowtest(cfg)
plt.title(cfg.flow)

# lamboseen_flow
cfg.flow = "cellular_flow" 
cfg.cellular_flow['vmax'] = 10
flowtest(cfg)
plt.title(cfg.flow)

plt.show()
```
