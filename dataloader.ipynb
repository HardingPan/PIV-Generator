````python
import os
import numpy as np
import torch
from torch.utils.data import Dataset

import matplotlib.pyplot as plt


class DatasetPIV(Dataset):
    def __init__(self, data_dir):
        self.data_dir = data_dir
        self.data_list = os.listdir(self.data_dir)
        
    def __len__(self):
        return len(self.data_list)

    def __getitem__(self, idx):
        data_path = os.path.join(self.data_dir, self.data_list[idx])
        data = np.load(data_path)
        img1, img2 = data['img1'], data['img2']
        u, v = data['u'], data['v']
        # if self.transform:
            # image = self.transform(image)
        return img1, img2, u, v
````

````python
cai_train_path = '/media/newdisk/datasets_piv/piv_cai/train'
cai_test_path = '/media/newdisk/datasets_piv/piv_cai/test'

cai_train_data =  DatasetPIV(cai_train_path)
cai_test_data =  DatasetPIV(cai_test_path)

figure = plt.figure(figsize=(8, 8))
cols, rows = 3, 3
for i in range(1, cols * rows + 1):
    sample_idx = torch.randint(len(cai_train_path), size=(1,)).item()
    img1, img2, u, v = cai_train_data[sample_idx]
    figure.add_subplot(rows, cols, i)
    # plt.title(labels_map[label])
    plt.axis("off")
    plt.imshow(img1.squeeze(), cmap="gray")
plt.show()
````
